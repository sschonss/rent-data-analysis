```python
import pandas as pd
```

```python
df = pd.read_csv('../aluguel.csv', sep=';')
```

```python
df
```

```python
classes = [0, 2, 4, 6, 100]
```

```python
quartos = pd.cut(df.Quartos, classes)
```

```python
quartos
```

```python
pd.value_counts(quartos)
```

```python
labels = ['0 e 2 quartos', '3 e 4 quartos', '5 e 6 quartos', '7 quartos ou mais']
```

```python
quartos = pd.cut(df.Quartos, classes, labels = labels)
```

```python
quartos
```

```python
pd.value_counts(quartos)
```

```python
quartos = pd.cut(df.Quartos, classes, labels = labels, include_lowest=True)
```

```python
pd.value_counts(quartos)
```
