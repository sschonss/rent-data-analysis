```python
import pandas as pd
```

```python
df = pd.read_csv('residencial.csv', sep=';')
```

```python
df
```

### Selecione somente os imóveis classificados com tipo 'Apartamento'.

```python
selecao_apt = df['Tipo'] == 'Apartamento'
n1 = df[selecao_apt].shape[0]
n1
```

### Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.

```python
selecao_casa = (df['Tipo'] == 'Casa') | (df['Tipo'] == 'Casa de Condomínio') | (df['Tipo'] == 'Casa de Vila')
n2 = df[selecao_casa].shape[0]
n2
```

### Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.

```python
selecao_area = (df['Area'] <= 100) & (df['Area'] >= 60)
n3 = df[selecao_area].shape[0]
n3
```

### Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.

```python
selecao_quartos = (df['Quartos'] >= 4) & (df['Valor'] < 2000)
n4 = df[selecao_quartos].shape[0]
n4
```

```python
print("Num de imóveis classificados com tipo 'Apartamento'. -> {}". format(n1))

print("Num de imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'. -> {}". format(n2))

print("Num de imóveis com área entre 60 e 100 metros quadrados, incluindo os limites. -> {}". format(n3))

print("Num de imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00. -> {}". format(n4))
```
